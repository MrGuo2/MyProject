```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function
import argparse
import sys
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets('.', one_hot=True)
```

```python
learn_rate = tf.placeholder(tf.float32)
# 更改初始化 添加隐层
X = tf.placeholder(tf.float32, [None, 784], name='X')
# 使用固定均值方差的权重初始化
# 不同方式的初始值会影响模型的准确度
# 固定值的效果最差，一般使用：Xvaier或MSRA初始化
W = tf.Variable(tf.truncated_normal(shape=[784, 200], mean=0, stddev=0.05), name='weight')
b = tf.Variable(tf.zeros([200]), name='bias')
W2 = tf.Variable(tf.truncated_normal(shape=[200,10]), name='weight2')
b2 = tf.Variable(tf.zeros([10]))
'''
在隐层增加神经元的个数，训练时间明显增长了，而且模型复杂度相比200个神经元的h1来说高了很多，但训练结果不是很好，产生了过拟合。
如果增加相同神经元个数的隐层，可以直接用循环来迭代增加隐层数量。
W = tf.Variable(tf.truncated_normal(shape=[784, 1000], mean=0, stddev=0.05), name='weight')
b = tf.Variable(tf.zeros([1000]), name='bias')
W2 = tf.Variable(tf.truncated_normal(shape=[1000,10], mean=0, stddev=0.05), name='weight2')
b2 = tf.Variable(tf.zeros([10]))
'''


# 在h1和输出之间添加一个矩阵接收hidden1激活后的结果并作为输入
# 提高隐层数量能够提高模型的准确性，但也增加了模型的复杂度容易产生过拟合
hidden1 = tf.nn.sigmoid(tf.matmul(X, W)+b)
logits = tf.matmul(hidden1, W2) + b2
# 此时的y还没有激活
y = tf.placeholder(tf.float32, [None, 10], name='y')

cross_entropy = tf.reduce_mean(
    tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=logits))
# 添加正则化防止模型产生过拟合
# l2正则化
l2_loss = tf.nn.l2_loss(W) + tf.nn.l2_loss(W2)
total_loss = cross_entropy + 2e-5*l2_loss

'''
函数training()通过梯度下降法为最小化损失函数增加了相关的优化操作，
在训练过程中，先实例化一个优化函数，比如 tf.train.GradientDescentOptimizer，并基于一定的学习率进行梯度优化训练
minimize()梯度下降
'''
train_step = tf.train.GradientDescentOptimizer(learn_rate).minimize(total_loss)

correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(logits, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

grap = tf.get_default_graph()
grap.as_graph_def()
```

```python
# tf.InteractiveSession()是一种交互式的session方式，它让自己成为了默认的session，
# 也就是说用户在不需要指明用哪个session运行的情况下，就可以运行起来，这就是默认的好处。这样的话就是run()和eval()函数可以不指明session啦。
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())

lr = 1.0
for step in range(3000):
    batch_X, batch_y = mnist.train.next_batch(100)
    if step<1800:
        lr = 1.2
    elif step<3400:
        lr = 0.3
    else:
        lr = 0.01
    _, loss, l2_loss_value, total_loss_value = sess.run(
                    [train_step, cross_entropy, l2_loss, total_loss],
                    feed_dict={X:batch_X, y:batch_y, learn_rate:lr})
    if(step + 1)%100 == 0:
        print('#' * 40)
        print('step %d, entropy loss: %f, l2_loss: %f, total loss: %f' % 
            (step+1, loss, l2_loss_value, total_loss_value))
        print(sess.run(accuracy, feed_dict={X:batch_X, y:batch_y}))
        print(sess.run(accuracy, feed_dict={X:mnist.test.images,
                                            y:mnist.test.labels}))
```
